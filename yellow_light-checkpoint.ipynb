```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
img = cv2.imread('yellow_light.jpg')  # Replace with the actual path to your image
img_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# Convert the image to HSV color space
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)


#upper_yellow  HSV = 60,100,100  
# H = 60*179/360=30
# S = 100* 255/100 = 255
# V = 100* 255/100 =  255
# Define the lower and upper bounds for the yellow color in HSV space
lower_yellow = np.array([20, 100, 100])
upper_yellow = np.array([30, 255, 255])  

# Create a mask for the yellow color
mask_yellow = cv2.inRange(img_hsv, lower_yellow, upper_yellow)

# Display the mask
plt.imshow(mask_yellow, cmap='gray')
plt.title('Yellow Color Mask')
plt.axis('off')
plt.show()
```

```python
fig,ax = plt.subplots(1,2, figsize = (10,10))
ax[0].imshow(img_RGB)
ax[1].imshow(mask_yellow, cmap = "gray")
plt.axis('off')
plt.show()
```
